```python
from pyspark.sql import SparkSession
import os

# Créer la session Spark
spark = SparkSession.builder \
    .appName("BDA-Price-Check") \
    .config("spark.executor.memory", "4g") \
    .config("spark.driver.memory", "4g") \
    .getOrCreate()

# Définir le chemin des données
prices_path = os.path.expanduser("~/bda-project/data/prices/*.csv")
print(f"Chemin des données : {prices_path}")

# Charger les données
try:
    df = spark.read.option("header", True) \
        .option("inferSchema", True) \
        .csv(prices_path)
    
    # Afficher le schéma
    print("\n Schéma des données :")
    df.printSchema()
    
    # Afficher les premières lignes
    print("\n Aperçu des données :")
    df.show(10, truncate=False)
    
    # Statistiques
    print(f"\n Nombre total de lignes : {df.count():,}")
    print(f" Nombre de colonnes : {len(df.columns)}")
    
    # Colonnes disponibles
    print(f"\n Colonnes disponibles : {df.columns}")
    
except Exception as e:
    print(f" Erreur : {e}")
finally:
    spark.stop()
```
